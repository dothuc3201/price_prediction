## The purpose of this notebook is to turn the datasets returned from the lag creator to be grouped by the day and add the features extracted for that day (average polarity and volume)

```python
#imports
import pandas as pd
```

```python
#get data with following lag
granularity = "days"
lag = 7
analyser = "TextBlob"
# analyser = "vader"
```

```python
#prepare filename to read
folder = "./../datasets/tweets_and_prices/"+analyser+"/";
filename = folder+"/final_data_lag_"+granularity+"_"+str(lag)+".csv" if (lag > 0) else folder+"/final_data_no_lag.csv"
```

```python
#read data
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
#remove extra columns
df = df.drop(['Unnamed: 0', 'Unnamed: 0_x',  'Unnamed: 0_y'], axis=1)
```

```python
#copy dataframe so that we do not have to stay reloading data after changes
df_copy = df.copy()
```

```python
df_copy.info()
```

```python
#get datetime
df_copy["dt"] = pd.to_datetime(df_copy['timestamp'])
df_copy = df_copy.drop(['timestamp'], axis=1)
```

```python
#floor datetimes to day
df_copy['DateTime'] = df_copy['dt'].dt.floor('d')
```

```python
df_copy['DateTime']
```

```python
#group tweet volumes daily
grouped_volumes = df_copy.groupby(pd.Grouper(key='dt',freq='d')).size().reset_index(name='tweet_vol')
```

```python
#set index to datetime
grouped_volumes.index = pd.to_datetime(grouped_volumes.index)
grouped_volumes = grouped_volumes.set_index('dt')
```

```python
#set type of volumes to float
grouped_volumes['tweet_vol'] = grouped_volumes['tweet_vol'].astype(float)
```

```python
grouped_volumes.head()
```

```python
#set datetime to index of dataframe
df_copy.index = pd.to_datetime(df_copy.index)
```

```python
#group by day and take the average of values in that day
df_copy = df_copy.groupby('DateTime').agg(lambda x: x.mean())
```

```python
df_copy.info()
```

```python
#set DateTime 
df['DateTime'] = df['dt'].dt.floor('d')
```
